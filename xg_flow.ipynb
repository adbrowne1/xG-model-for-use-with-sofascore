```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import themepy
theme = themepy.Theme()
```

### This is a sample visualisation that uses the data from the game in the predictor script to plot the cumulative xG for each team throughout the course of the game ####

```python
# From the xG_predictor script the xG header has been added to the csv
df = pd.read_csv('/Users/adambrowne/Desktop/Personal /LOI Project/Data/Pats_vs_Shels_2023.csv')
df
```

```python
a_xg = []
h_xg = []
a_min = []
h_min = []

# find home & away team
h_team = df['teamid'].iloc[0]
a_team = df['teamid'].iloc[16]

for x in range(len(df['xG'])):
    if df['teamid'][x]== h_team:
        h_xg.append(df['xG'][x])
        h_min.append(df['time'][x])
        


for x in range(len(df['xG'])):
    if df['teamid'][x]== a_team:
        a_xg.append(df['xG'][x])
        a_min.append(df['time'][x])
print(h_min)
print(a_min)
```

```python
a_cumulative = np.cumsum(a_xg)
h_cumulative = np.cumsum(h_xg)
print(h_cumulative)


a_min.sort()
h_min.sort()

print(a_min)


a_min = np.append(0,a_min)
a_min = np.append(a_min,90)
h_min = np.append(0,h_min)
h_min = np.append(h_min,90)

a_cumulative=np.append(0,a_cumulative)
a_cumulative=np.append(a_cumulative,1.02)

print(a_cumulative)
h_cumulative=np.append(0,h_cumulative)
h_cumulative=np.append(h_cumulative,4.56)
print(h_cumulative)
```

```python

# can adjust themes and params here
(theme
 .set_theme('sample-dark')
 .set_font("Arial")
)
fig, ax = plt.subplots( figsize = (11,5) )
plt.plot(a_min, a_cumulative, color = 'red', label = 'St Patricks Athletic' )
plt.plot(h_min, h_cumulative, color = 'White', label = 'Shelbourne')
plt.xlabel('Minute')
plt.ylabel('xG')
plt.title('Shelbourne vs St Patricks Athletic 01/09\n xG Flow')
plt.legend(loc = 'best')
plt.show()
```

### The outcome of this match was a 3-2 with for shelbourne from two late goals. The xG flow chart shows that although the goals came late, shelbourne consistently created high quality chances throughout the game. ###
